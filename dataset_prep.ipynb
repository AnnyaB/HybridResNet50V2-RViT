```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

sample_paths = list(Path("data/processed/train/glioma").glob("*.jpg"))[:4]

fig, axes = plt.subplots(1, len(sample_paths), figsize=(10, 3))
for ax, p in zip(axes, sample_paths):
    img = Image.open(p)
    ax.imshow(img)
    ax.set_title("glioma (train)")
    ax.axis("off")

plt.tight_layout()
```

```python
import pandas as pd

stats = pd.read_csv("results/raw_image_stats.csv")

good_mask = (
    (stats["width"] == 512)
    & (stats["height"] == 512)
    & (~stats["suspect"])
)
weird_mask = (
    (stats["width"] != 512)
    | (stats["height"] != 512)
    | (stats["suspect"])
)

print("Good candidates sample:")
print(stats[good_mask][["orig_path", "width", "height", "suspect"]].head())

print("\nWeird candidates sample:")
print(stats[weird_mask][["orig_path", "width", "height", "suspect"]].head())
```
